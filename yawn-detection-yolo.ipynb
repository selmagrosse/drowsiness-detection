```python
import os

DATASET_PATH = '/kaggle/input/yawdd-photos/yawdd'

# Create data.yaml
data_yaml = f"""
train: {DATASET_PATH}/images/train
val: {DATASET_PATH}/images/val

nc: 1
names: 
    0: yawn
"""

with open("yawdd.yaml", "w") as f:
    f.write(data_yaml)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt") 
model.train(data="yawdd.yaml", epochs=100, imgsz=640)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('runs/detect/train4/results.png')
plt.figure(figsize=(10,6))
plt.imshow(img)
plt.axis('off')
plt.title('YOLOv8 Training Metrics')
plt.show()
```
